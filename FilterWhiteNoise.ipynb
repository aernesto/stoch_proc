For the beginning of this notebook, I would like to confirm equation 6.2.30 from "Stochastic Differential Equations", by Oksendal.

In Example 6.2.9, the author defines:
- The system (I will take white noise)
$$dX_t=0, X_t=X_0; E[X_0]=\hat{X_0}, E[X_0^2]=a^2$$
- The observations ($W_t$ and $V_t$ are independent Brownian motion)
$$H_t=X_t+mW_t; dZ_t=X_tdt+mdV_t; Z_0=0$$
- The best estimate in the $L^2$ sense:
$$\hat{X}_t=\frac{m^2}{m^2+a^2t}\hat{X}_0 + \frac{a^2}{m^2+a^2t}Z_t; t\geq0$$

I would like to run several simulations to observe the behavior of this estimate, and the corresponding error.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

First, let's simulate a single trajectory of $X_t$

```python
# X_t in the equations above
class WhiteNoise:
    def __init__(self, mean_value, stdev_value, horizon=1, timestep=1/1000):
        self.mean_value = mean_value
        self.stdev_value = stdev_value
        self.horizon = horizon
        self.timestep = timestep
        self.num_samples = np.floor(self.horizon / self.timestep).astype(int)
        
    def sample_white_noise(self):
        return np.random.normal(self.mean_value, self.stdev_value, self.num_samples.astype(int))
```

```python
initial_time = 0
final_time = 1
num_time_steps = 1000
time_step_size = (final_time - initial_time) / num_time_steps
time_points = np.linspace(initial_time, final_time, num_time_steps)
```

```python
plt.plot(time_points, WhiteNoise(1, 1, final_time - initial_time, time_step_size).sample_white_noise())
plt.title('Samples from white noise process')
```
